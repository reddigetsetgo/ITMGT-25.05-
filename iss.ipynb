```python
import matplotlib.pyplot as plt
import time
import requests

timestamps = []
longitudes = []
latitudes = []

for record in range(100):
    
    data = []
    res = requests.get('http://api.open-notify.org/iss-now.json')
    json_data = res.json()
    
    # Data Points
    data.append(json_data['timestamp'])
    data.append(float(json_data['iss_position']['longitude']))
    data.append(float(json_data['iss_position']['latitude']))
    print(data)
    
    # Path of ISS
    timestamps.append(json_data['timestamp'])
    longitudes.append(float(json_data['iss_position']['longitude']))
    latitudes.append(float(json_data['iss_position']['latitude']))
    
    time.sleep(10)
```

```python
plt.plot(longitudes, latitudes, '-o')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
